### Filtrado de Datos

```python
# Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50.

# Como no existe la columna Ventas, se utiliza Quantity en vez
df[df["Total Amount"] > 50]
```

```python
# Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5.

# Como no existe la columna Precio, se usa una columna equivalente denominada Price Per Unit.
df[df["Price per Unit"] < 0.5]

# Notar que el DataSet no contiene productos con un precio menor a 0.5, por lo que se
# busca el menor precio y se intenta con un valor un poco mayor en vez.

print(df.min())
# 25 es el minio precio por unidad, por lo que probamos con 26
df[df["Price per Unit"] < 26]

# De los datos se puede destacar que una gran mayoría de los productos vendidos tenían un precio menor incluso a 26
```

```python
# De los datos se puede notar que una gran mayoría de productos poseen un precio menor a 26.
# Para determinar la cantidad total de filas, o se saca el len, o el shape de los datos filtrados

cant_filas_len = len(df[df["Price per Unit"] < 26])
cant_filas_shape, _ = df[df["Price per Unit"] < 26].shape

print(f"Cant. Filas con precio menor a 26 usando len(): {cant_filas_len}")
print(f"Cant. Filas con precio menor a 26 usando shape: {cant_filas_shape}")
```

```python
# Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30.

# Para ver si es posible buscar Manzanas en el dataset, primero verificamos si existe alguna columna que cumpla este criterio
cant_productos_manzana = len(df[df["Product Category"] == "Manzanas"])
print(f"Cantidad de productos con tipo manzana: {cant_productos_manzana}")
```

```python
# Como no existe, entonces buscamos una que si esté en el data set
df.head(5)
```

```python
# Seleccionamos la categoria Clothing, y usamos la columna Total Amount para obtener las ventas
# Notar que para usar nombres de columnas con espacios en un .query(), es necesario hacerlo así: `Columna Espacios`
df.query("`Product Category` == 'Clothing' & `Total Amount` > 30")
```
